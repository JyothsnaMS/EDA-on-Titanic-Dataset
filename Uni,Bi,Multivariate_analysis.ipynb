# Univariate Analysis - is classification possible with a single attribut

```python
df_setosa = df.loc[df['species'] == 'setosa']
```

```python
df_setosa
```

```python
df_virginica = df.loc[df['species'] == 'virginica']
df_versicolor = df.loc[df['species'] == 'versicolor']
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']), 'o') #x-value.
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']), 'o')
#since univariate, y is made 0
plt.xlabel('Petal length')
plt.show()
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']), 'o')
plt.xlabel('Petal length')
plt.show()
```

# Bivariate analysis-2 features

```python
sns.FacetGrid(df, hue="species", size=5).map(plt.scatter, "sepal_length", "sepal_width").add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue="species", size=5).map(plt.scatter, "sepal_length", "sepal_width")
plt.show()
```

```python
sns.FacetGrid(df, hue="species", size=5).map(plt.scatter, "petal_length", "petal_width").add_legend()
plt.show()
```
